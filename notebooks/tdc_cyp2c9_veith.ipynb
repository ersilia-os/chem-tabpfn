```python
from tdc.single_pred import ADME
data = ADME(name='CYP2C9_Veith')
split = data.get_split()
```

```python
print(split['train'].shape)
print(split['test'].shape)
print(split['valid'].shape)
```

```python
from eosce.models import ErsiliaCompoundEmbeddings
```

```python
descriptor = ErsiliaCompoundEmbeddings()
```

```python
from operator import itemgetter

train, test, valid = itemgetter('train', 'test', 'valid')(split)
train.head()
```

```python
train_transformed = descriptor.transform(train.Drug)
```

```python
train_transformed.shape
```

```python
test_transformed = descriptor.transform(test.Drug)
test_transformed.shape
```

```python
from ensemble_tabpfn import EnsembleTabPFN
```

```python
clf = EnsembleTabPFN()

clf.fit(train_transformed, train.Y)
```

```python
preds = clf.predict(test_transformed)
```

```python
preds.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(test.Y, preds))
```

```python
clf_cluster = EnsembleTabPFN(feature_sampler="cluster")
clf_cluster.fit(train_transformed, train.Y)
```

```python
preds_cluster = clf_cluster.predict(test_transformed)
print(accuracy_score(test.Y, preds_cluster))
```

```python
clf_pca = EnsembleTabPFN(feature_sampler="pca")
clf_pca.fit(train_transformed, train.Y)
preds_pca = clf_pca.predict(test_transformed)
print(accuracy_score(test.Y, preds_pca))
```

```python
clf_lrp = EnsembleTabPFN(feature_sampler="lrp")
clf_lrp.fit(train_transformed, train.Y)
preds_lrp = clf_pca.predict(test_transformed)
print(accuracy_score(test.Y, preds_lrp))
```

```python
clf_random = EnsembleTabPFN(feature_sampler="random")
clf_random.fit(train_transformed, train.Y)
preds_rand = clf_random.predict(test_transformed)
print(accuracy_score(test.Y, preds_rand))
```
